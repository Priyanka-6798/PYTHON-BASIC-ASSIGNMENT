## PYTHON BASIC ASSIGNMENT

1. What is Python, and why is it popular?

   Ans: Python is a high-levelprogramming language known for its simplicity and readability. It's popular because of its easy-to-learn syntax, vast standard library, strong community support, and versatility used in web development, data science, AI, automation.

2. What is an interpreter in Python?

   Ans: An interpreter in Python is a program that reads and executes Python code line by line. It converts human-readable Python code into machine-readable without a separate step.

3. What are pre-defined keywords in Python?

   Ans: Pre-defined keywords in Python are reserved words that have specific meanings and purposes in the language. Examples include if, else, for, while, def, class, import, True, False, None.

4. Can keywords be used as variable names?

   Ans: No, keywords cannot be used as variable names because they are reserved for specific language functionality. Attempting to use them as variables will result in a syntax error.

5. What is mutability in Python?

   Ans: Mutability refers to whether an object's value can be changed after it's created. Mutable objects (like lists) can be modified, while immutable objects (like strings) cannot be changed after creation.
6. Why are lists mutable, but tuples are immutable?

   Ans: Lists are mutable because they are designed to be modified (add, remove, change elements). Strings are immutable because they are meant to represent fixed collections of items where the sequence and values shouldn't change, which provides safety and can be used as dictionary keys.

7. What is the difference between "==" and "is" operators in Python?

   Ans: "==" compares the values of two objects, while "is" checks if two variables refer to the exact same object in memory.

8. What are logical operators in Python?

   Ans: Logical operators in Python are and, or, and not. They are used to combine conditional statements:

    and: True if both operands are true
    or: True if at least one operand is true
    not: Inverts the boolean value

9. What is type casting in Python?

   Ans: Type casting is the process of converting a value from one data type to another, such as converting a string to an integer or a float to a string.

10. What is the difference between implicit and explicit type casting?

   Ans: Implicit type casting is automatically performed by Python (e.g., adding an integer and float results in a float). Explicit type casting is done manually by the programmer using functions like int(), float(), str(), etc.    

11. What is the purpose of conditional statements in Python?

    Ans: Conditional statements (if, elif, else) allow the program to execute different blocks of code based on whether certain conditions are true or false, enabling decision-making in programs.

12. How does the elif statement work?

    Ans: elif (short for "else if") is used to check multiple conditions after an initial if statement. It provides an alternative condition to check if the previous conditions were false, and comes before the final else clause.

13. What is the difference between for and while loops?

    Ans: A for loop iterates over a sequence (like a list, tuple, or string) or other iterable objects, executing the block for each item. A while loop repeatedly executes a block of code as long as a condition remains true.

14. Describe a scenario where a while loop is more suitable than a for loop.

    Ans: A while loop is more suitable when you don't know in advance how many times you need to loop, such as when reading input from a user until they enter a specific value (e.g., "quit"), or when waiting for a certain condition to become true.

10. Write a code to demonstrate type casting with list elements.

    Ans: # Demonstration of type casting with list elements

# Mixed list with different data types
mixed_list = ['10', 20, 30.5, '40.5', True, 'Hello', '0', '-15', False, None]

print("Original list:", mixed_list)

# 1. Convert elements to integers (where possible)
int_list = []
for item in mixed_list:
    try:
        # Handle boolean values first (since bool is subclass of int)
        if isinstance(item, bool):
            int_list.append(int(item))
        elif item is None:
            continue  # Skip None values
        else:
            int_list.append(int(float(item)))  # Convert to float first to handle strings like '40.5'
    except (ValueError, TypeError):
        pass

print("\nElements successfully converted to integers:", int_list)

# 2. Convert elements to floats (where possible)
float_list = []
for item in mixed_list:
    try:
        if isinstance(item, bool):
            float_list.append(float(item))
        elif item is None:
            continue
        else:
            float_list.append(float(item))
    except (ValueError, TypeError):
        pass

print("\nElements successfully converted to floats:", float_list)

# 3. Convert elements to strings
str_list = [str(item) for item in mixed_list]
print("\nAll elements converted to strings:", str_list)
